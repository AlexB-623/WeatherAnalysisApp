```python
import openmeteo_requests
import requests_cache, pandas as pd
from retry_requests import retry
import requests
from call_for_data import make_master_dataframe
```

```python
from call_test import latitude
# Setup the Open-Meteo API client with cache and retry on error
cache_session = requests_cache.CachedSession('.cache', expire_after = -1)
retry_session = retry(cache_session, retries = 5, backoff_factor = 0.2)
open_meteo = openmeteo_requests.Client(session = retry_session)
gps = (38.29088, -72.61076)
latitude = gps[0]
longitude = gps[1]
start_date = "1988-06-23"
end_date ="1988-06-24"
```

```python
url = "https://archive-api.open-meteo.com/v1/archive"
params = {
		"latitude": latitude,
		"longitude": longitude,
		"start_date": start_date,
		"end_date": end_date,
		"hourly": ["temperature_2m", "precipitation", "cloud_cover", "wind_speed_100m"],
		"timezone": "auto",
        "temperature_unit": "fahrenheit"
	}
responses = open_meteo.weather_api(url, params=params)
```
